```python
"""DNA SEQUENCE ANALYZER=>
A program that:

1)Takes DNA sequence

Shows:
-Length
-GC %
-Complement
-Reverse complement
-RNA sequence
-Protein translation"""
from Bio.Seq import Seq
seq = input("enter your DNA sequence: ")
gc = (seq.count("G")+seq.count(""))/len(seq)*100
print("length of the DNA sequence = ",len(seq))
print(f"GC content in DNA is {gc}% .:2f")
print(f"Complement of a DNA Sequence is  ",seq.complement())
print(f" Reverse Complement of a DNA Sequence is  ",seq.reverse_complement())
print("RNA sequence is",seq.transcribe())
print("protein chain is ",seq.translate())
```
